```python
import pandas as pd
import numpy as np
import sys
import sweetviz as sv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#!{sys.executable} -m pip install sweetviz
```

```python
df = pd.read_csv('SFAirbnb.csv')
```

```python
df.columns.values
```

```python
#Drop Unimportant features
features = ['host_identity_verified', 'neighbourhood_cleansed', 'zipcode', 'property_type', 'room_type', 'accommodates', 'bathrooms', 'bedrooms', 'beds', 'security_deposit', 'instant_bookable', 'price']
```

```python
df = df[features]
```

```python
df
```

```python
df["security_deposit"] = df["security_deposit"].fillna(df['security_deposit'].mode()[0])
```

```python
df["bathrooms"] = df["bathrooms"].fillna(df['bathrooms'].mode()[0])
```

```python
df["bedrooms"] = df["bedrooms"].fillna(df['bedrooms'].mode()[0])
```

```python
df["beds"] = df["beds"].fillna(df['beds'].mode()[0])
```

```python
df["host_identity_verified"] = df["host_identity_verified"].fillna(df['host_identity_verified'].mode()[0])
```

```python
df['zipcode'].isnull() == True
```

```python
df = df.dropna()
```

```python
df = df.reset_index()
```

```python
df = df.drop('zipcode', axis=1)
```

```python
df
```

```python
#my_report = sv.analyze(df)
```

```python
#my_report.show_html()
```

```python
df['security_deposit'] = df['security_deposit'].str.replace(r'\D', '').astype(int)
```

```python
df["price"] = df["price"].str.replace('$', '')
df["price"] = df["price"].str.replace(',', '')
```

```python
df['price'] = df['price'].astype(float)
```

```python
df
```

```python
df_dummies = pd.get_dummies(df)
```

```python
df_dummies.drop('index', axis=1, inplace = True)
```

```python
dummies_report = sv.analyze(df_dummies)
dummies_report.show_html()
```

```python
sns.distplot(df['price'])
```

```python
df_dummies = df_dummies[df_dummies.price != 0]
```

```python
target = np.log(df_dummies['price'])
```
